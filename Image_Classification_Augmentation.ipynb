# Image Classification with Data Augmentation

This notebook regenerates the flower vs. landscape patch dataset, trains both the baseline and augmented CNNs, saves metrics to `outputs/`, and surfaces the recorded validation accuracies.

```python
from pprint import pprint
from train import DatasetConfig, TrainingConfig, run_all_experiments

dataset_config = DatasetConfig(seed=42)
training_config = TrainingConfig(
    epochs=10,
    batch_size=32,
    learning_rate=1e-3,
    output_dir="outputs",
)

results = run_all_experiments(dataset_config=dataset_config, training_config=training_config)
pprint({run: info['best_val_accuracy'] for run, info in results.items()})
```

```python
import json
import os
from IPython.display import display, Markdown

def load_history(path):
    with open(path, 'r', encoding='utf-8') as f:
        return json.load(f)

rows = []
for run_name, info in results.items():
    history = load_history(info['metrics_path'])
    for entry in history:
        rows.append((run_name, entry['epoch'], entry['train_accuracy'], entry['val_accuracy']))

lines = ["| Run | Epoch | Train Accuracy | Val Accuracy |", "| --- | --- | --- | --- |"]
for run, epoch, train_acc, val_acc in rows:
    lines.append(f"| {run} | {epoch} | {train_acc:.3f} | {val_acc:.3f} |")
display(Markdown('
'.join(lines)))

print("Artefacts saved to:", os.path.abspath(training_config.output_dir))
```
